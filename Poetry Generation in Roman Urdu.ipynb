```python
URDU_DIACRITICS  = ['ِ', 'ٰ', 'ُ', 'ٍ', 'ً', 'َ']
URDU_DIGIT = ['۶', '۴', '۵', '۸', '۲', '۰', '۷', '۹', '۳', '۱']
URDU_PUNCTUATIONS = ['\200F','\n', '\u200f','؛', '۔', '٫', '٪', '؟', '،', ')', '(', '{', '}']
URDU_EXTRA_CHARACTER = [1623,1538, 1617,1537,1551,1539, 1624, 1550,1622,1644,1555,1620,1619,1552,1554,1556,1612,1557,1549,1553,1618,1536]

def removing_diacritics(text):
    for i in text:
        if i in URDU_DIACRITICS:
            text = text.replace(i, "")
    return text


def remove_numbers(text):
    for i in text:
        if i in URDU_DIGIT:
            text = text.replace(i, "")
    return text


def removing_extra_character(text):
    for i in text:
        if ord(i) in URDU_EXTRA_CHARACTER:
            text = text.replace(i, "")
    return text


def removing_punctuations(text):
    for i in text:
        if i in URDU_PUNCTUATIONS:
            text = text.replace(i, "")
    return text


def remove_newline_tab_character(text):
    regex = re.compile(r"[\n\r\t]")
    return regex.sub("", text)
    

unicode_mapping = {
            u"\u0627":"A",
            u"\u0627":"A", 
            u"\u0675":"A", 
            u"\u0673":"A", 
            u"\u0630":"A", 
            u"\u0622":"AA", 
            u"\u0628":"B", 
            u"\u067E":"P", 
            u"\u062A":"T", 
            u"\u0637":"T", 
            u"\u0679":"T", 
            u"\u062C":"J", 
            u"\u0633":"S", 
            u"\u062B":"S", 
            u"\u0635":"S", 
            u"\u0686":"CH", 
            u"\u062D":"H", 
            u"\u0647":"H", 
            u"\u0629":"H", 
            u"\u06DF":"H", 
            u"\u062E":"KH", 
            u"\u062F":"D", 
            u"\u0688":"D", 
            u"\u0630":"Z", 
            u"\u0632":"Z", 
            u"\u0636":"Z", 
            u"\u0638":"Z", 
            u"\u068E":"Z", 
            u"\u0631":"R", 
            u"\u0691":"R", 
            u"\u0634":"SH", 
            u"\u063A":"GH", 
            u"\u0641":"F", 
            u"\u06A9":"K", 
            u"\u0642":"K", 
            u"\u06AF":"G", 
            u"\u0644":"L", 
            u"\u0645":"M", 
            u"\u0646":"N", 
            u"\u06BA":"N", 
            u"\u0648":"O", 
            u"\u0649":"Y", 
            u"\u0626":"Y", 
            u"\u06CC":"Y", 
            
            u"\u06D2":"E", 
            u"\u06C1":"H",
            u"\u064A":"E"  ,
            u"\u06C2":"AH"  ,
            u"\u06BE":"H"  ,
            u"\u0639":"A"  ,
            u"\u0643":"K" ,
            u"\u0621":"A",
            u"\u0624":"O",
            u"\ufe85":"O",
            u"\u060C":"" #seperator ulta comma
}

def TranslateString(string, reverse=0):
    
    for k, v in unicode_mapping.items():
        if not reverse:
            string = string.replace(k, v)
        else:
            string = string.replace(v, k)

    return string


def Urdu_to_RomanUrdu(word):
    #Remove Urdu diacritics
    word=removing_diacritics(word)
   # print('->  Remove Urdu diacritics')
   # print(word)
    
    #Remove urdu number e.g '۶', '۴', '۵', '۸', '۲', '۰', '۷', '۹', '۳', '۱
    word=remove_numbers(word)
   # print('->  urdu number')
   # print(word)
    
    word=removing_extra_character(word)
    #Remove urdu punctuations 
    word=removing_punctuations(word)
   # print('->  Remove urdu punctuations')
   # print(word)
    
    #Remove '\t','\n'
    word=remove_newline_tab_character(word)
    
    print(TranslateString(word))
   
```

# Bigram Model

```python
bigram_model = defaultdict(lambda: defaultdict(lambda: 0))

def bigram_probability():
    

    #
    bigrams = ngrams(text.split(), 2)
    #count frequency of word 
    for w1, w2 in bigrams:
        bigram_model[(w1)][w2] += 1
        
    #calculate probability
    
    for w1 in bigram_model:
        total_count = float(sum(bigram_model[w1].values()))
        for w3 in bigram_model[w1]:
            bigram_model[w1][w3] /= total_count

            
def predict_bigram_word(w1):
    if dict(bigram_model[w1])=='':
        return ""
    else :
        return max(dict(bigram_model[w1]))

            



 
```

# Trigram Model

```python
model = defaultdict(lambda: defaultdict(lambda: 0))
def trigram_probability():
    trigrams = ngrams(text.split(), 3)
    

    for w1, w2, w3 in trigrams:
        model[(w1, w2)][w3] += 1
        
    for w1_w2 in model:
        total_count = float(sum(model[w1_w2].values()))
        for w3 in model[w1_w2]:
            model[w1_w2][w3] /= total_count
            
def predict_trigram_word(w1,w2):
    return max(dict(model[w1,w2]))
```

# Calculate Laplace smoothing

```python
def generate_ngrams(text, n_gram=1, stop=True):
    
#     stop = set(stopwords.words("english")) if stop else {}

    token = [
        token for token in text.lower().split(" ") #if token != "" if token not in stop
    ]
    z = zip(*[token[i:] for i in range(n_gram)])
    ngrams = [" ".join(ngram) for ngram in z]

    return ngrams
```

```python
#count unigram word
unigrams = defaultdict(int)
total_word=0
for word in generate_ngrams(text,1):
            unigrams[word] += 1
```
